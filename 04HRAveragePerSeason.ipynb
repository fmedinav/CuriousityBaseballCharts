```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters used in the code. Change as needed before running the code
DATA_BASE_PATH = 'https://raw.githubusercontent.com/chadwickbureau/baseballdatabank/master/core/Batting.csv'
START_YEAR = 1871
END_YEAR = 2022
COLUMNS_TO_RETRIEVE = ['playerID', 'yearID', 'HR']
```

```python
# Read the data base
battingDf = pd.read_csv(DATA_BASE_PATH)
```

```python
# Allocate memory in a numpy array
battingArray = np.zeros([END_YEAR - START_YEAR, 2])
```

```python
# Populate the array with the desired averages
for i in range(START_YEAR, END_YEAR):
    condition1 = battingDf.yearID == i
    yearHRMean = battingDf.loc[condition1, :].HR.mean()
    battingArray[i - START_YEAR, 0] = i
    battingArray[i - START_YEAR, 1] = yearHRMean
```

```python
# Plot
plt.plot(battingArray[:,0], battingArray[:,1])
plt.title('HR Average per player per season')
plt.xlabel('Year')
plt.ylabel('Average HR per player')
plt.legend(['MLB Batting data base'])
plt.xlim([START_YEAR, END_YEAR])
plt.grid()
plt.show()
```
